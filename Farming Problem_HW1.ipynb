```python
import numpy as np
import gurobipy as gp
#testing git
```

```python
fert_levels=list(range(200,2300,100))
```

```python
import pandas as pd
optimal_wheat=[]
optimal_corn=[]
max_profit=[]
for fertilizer in fert_levels:
    farmMod=gp.Model()
    farmModx=farmMod.addMVar(2)
    farmMod.setObjective(2*farmModx[0]+3*farmModx[1],sense=gp.GRB.MAXIMIZE)
    conlist=[0]*3
    conlist[0]=farmMod.addConstr(farmModx[0]+farmModx[1]<=450)
    conlist[1]=farmMod.addConstr(3*farmModx[0]+2*farmModx[1]<=1000)
    conlist[2]=farmMod.addConstr(2*farmModx[0]+4*farmModx[1]<=fertilizer)
    farmMod.Params.OutputFlag=0
    farmMod.optimize()
    optimal_wheat.append(farmModx[0].x)
    optimal_corn.append(farmModx[1].x)
    max_profit.append(farmMod.ObjVal)
df=pd.DataFrame({"Fertilizer":fert_levels,
                 "Wheat":optimal_wheat,
                 "Corn":optimal_corn,
                 "Max Profit":max_profit})
```
